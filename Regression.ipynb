```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pycaret.datasets import get_data
```

```python
df=get_data('insurance')
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
from pycaret.regression import *
```

```python
reg=setup(data=df,target='charges')
```

```python
check_model = compare_models(include=['lr', 'dt', 'rf', 'xgboost', 'svm'], fold=4)

# Display the best model
print(check_model)
```

```python
reg1 = setup(df, target='charges', train_size=0.8, session_id=147,
             log_experiment=False, transformation=True, pca=True,
             pca_method='linear', pca_components=0.8, experiment_name='ins1')
```

```python
best_model = compare_models(exclude=['catboost'], fold=4)
```

```python
models()
```

```python
grb = create_model('gbr',fold=4)
```

```python
rf=create_model('rf',fold=4)
```

```python
lgbm=create_model('lightgbm',fold=4)
```

```python
tune_grb=tune_model(grb,n_iter=10,optimize='MAE')
```

Tune Parameters

```python
tune_grb
```

```python
tune_lgbm=tune_model(lgbm,n_iter=10,optimize='R2')
```

```python
tune_lgbm
```

```python
stacking_grb=stack_models(estimator_list=[grb,rf,lgbm],meta_model=tune_lgbm)
```

```python
en_lgbm=ensemble_model(lgbm,method='Boosting')
```

```python
blender=blend_models(estimator_list=[grb,rf,lgbm],fold=4)
```
